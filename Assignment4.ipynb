```python
class Solution:
    def minCut(self, s: str) -> int:
      return self.findMinLength(s[:len(s)], 0)

    def findMinLength(self, s, length):
      str_length = len(s)
      if self.check_palindrom(s) == False:
        length += 1
        return min(self.findMinLength(s[:str_length-1], length), self.findMinLength(s[1:str_length], length))
      else:
        return length

    def check_palindrom(self, s: str):
      print(s)
      str_length = len(s)
      if str_length == 1:
        return True
      start_index = 0
      end_index = str_length - 1
      while start_index < str_length/2:
        if s[start_index] != s[end_index]:
          return False
        start_index += 1
        end_index -= 1
      return True
```

```python
s = Solution()
print(s.minCut("cabababcbc"))
```
